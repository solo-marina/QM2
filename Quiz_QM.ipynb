```python
file_path = '/content/drive/My Drive/Colab Notebooks/Data_quiz (6).csv'
df = pd.read_csv(file_path)
print(df)
```

```python
def most_useful_languages(country, occupation):
    filtered_data = df[(df['Country'] == country) & (df[occupation].notnull())]

    if filtered_data.empty:
        return f"No data available for {country} and {occupation}."

    sorted_languages = filtered_data[['Language', occupation]].sort_values(by=occupation, ascending=False)

    result_string = f"For {country} and {occupation}: "

    for i in range(len(sorted_languages)):
      if i + 1 == 1:
        x = str(i + 1)+'st'
      if i + 1 == 2:
        x = str(i + 1)+'nd'
      if i + 1 == 3:
        x = str(i + 1)+'rd'
      result_string += f"The {x} most useful language would be {sorted_languages.iloc[i]['Language']}. "


    return result_string

unique_countries = df['Country'].unique()
unique_occupations = [col for col in df.columns if col not in ['Country', 'Language']]

for country in unique_countries:
    for occupation in unique_occupations:
        result = most_useful_languages(country, occupation)
        print(result)
```
